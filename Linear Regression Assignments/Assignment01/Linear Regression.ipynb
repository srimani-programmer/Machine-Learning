## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
dataset = df.iloc[:,:].values
dataset
```

## Data Preprocessing

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN',strategy='mean',axis=0)
imputer = imputer.fit(dataset[:,0:2])
dataset[:,0:2] = imputer.transform(dataset[:,0:2])
```

```python
X_ind = dataset[:,0]
Y_dep = dataset[:,1]
```

```python
X_ind = np.array(X_ind)
X_ind = X_ind.reshape(-1,1)
X_ind
```

```python
Y_dep
```

## Building a Machine Learning Model

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_ind,Y_dep)
```

## Importing the Test Datasets

```python
testdata = pd.read_csv('test.csv')
```

```python
testcleaning = testdata.iloc[:,:].values
```

# Preprocessing the Test Dataset

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN',strategy='mean',axis=0)
imputer = imputer.fit(testcleaning[:,0:2])
testcleaning[:,0:2] = imputer.transform(testcleaning[:,0:2])
```

```python
X_ind_test = testcleaning[:,0]
X_ind_test = np.array(X_ind_test)
X_ind_test = X_ind_test.reshape(-1,1)
X_ind_test
```

```python
Y_dep_test = testcleaning[:,1]
Y_dep_test
```

## Predicting the testSet Results 

```python
y_predict = regression.predict(X_ind_test)
```

## Visualising the training set results

```python
plt.scatter(X_ind,Y_dep,color = 'red')
plt.plot(X_ind,regression.predict(X_ind))
plt.title('Simple Linear Regression Model (Trained Model)')
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.show()
```

## Visualising the test set results

```python
plt.scatter(X_ind_test,Y_dep_test,color = 'red')
plt.plot(X_ind,regression.predict(X_ind))
plt.title('Simple Linear Regression Model (Trained Model)')
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.show()
```

## Calculating the accuracy of the TestData

```python
test_accuracy = regression.score(X_ind_test,Y_dep_test)
test_accuracy
```

```python
print(test_accuracy*100)
```

## Calculating the accuracy of the Trained Data

```python
train_accuracy = regression.score(X_ind,Y_dep)
train_accuracy
```
